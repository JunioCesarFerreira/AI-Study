# Distribuições de Probabilidades

Neste notebook estão alguns blocos de código para testes e geração de gráficos didáticos sobre distribuição de probabilidades.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom, norm

# Configuração da Distribuição Binomial
n, p = 10, 0.5  # Número de ensaios, Probabilidade de sucesso
x_binom = np.arange(0, n+1)
binom_probs = binom.pmf(x_binom, n, p)

# Configuração da Distribuição Normal
mu, sigma = 0, 1  # Média e desvio padrão
x_norm = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
norm_pdf = norm.pdf(x_norm, mu, sigma)

# Criando a figura e os eixos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Gráfico de Barras para Distribuição Binomial
ax1.bar(x_binom, binom_probs, color='blue', alpha=0.7)
ax1.set_title('Distribuição Binomial')
ax1.set_xlabel('Número de Sucessos')
ax1.set_ylabel('Probabilidade')
ax1.grid(True)

# Gráfico de Linha para Distribuição Normal
ax2.plot(x_norm, norm_pdf, color='red')
ax2.fill_between(x_norm, norm_pdf, alpha=0.3, color='red')
ax2.set_title('Distribuição Normal')
ax2.set_xlabel('Valor')
ax2.set_ylabel('Densidade de Probabilidade')
ax2.grid(True)

# Ajustando a disposição dos subplots
plt.tight_layout()

# Salvando a figura
plt.savefig('distribuicoes_probabilidades.png', dpi=300)

# Exibindo a figura
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom, poisson, multinomial

# Configuração da Distribuição Binomial
n, p = 10, 0.5
x_binom = np.arange(0, n+1)
binom_probs = binom.pmf(x_binom, n, p)

# Configuração da Distribuição de Poisson
mu_poisson = 3
x_poisson = np.arange(0, 10)
poisson_probs = poisson.pmf(x_poisson, mu_poisson)

# Configuração da Distribuição Multinomial
n_multinomial = 10
p_multinomial = [0.2, 0.3, 0.5]
x_multinomial = np.arange(0, n_multinomial+1)
multinomial_probs = multinomial.pmf([3, 3, 4], n_multinomial, p_multinomial)

# Criando a figura e os eixos
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))

# Gráfico de Barras para Distribuição Binomial
ax1.bar(x_binom, binom_probs, color='blue', alpha=0.7)
ax1.set_title('Distribuição Binomial')
ax1.set_xlabel('Número de Sucessos')
ax1.set_ylabel('Probabilidade')
ax1.grid(True)

# Gráfico de Linha para Distribuição de Poisson
ax2.plot(x_poisson, poisson_probs, 'bo-', markersize=8)
ax2.set_title('Distribuição de Poisson')
ax2.set_xlabel('Número de Eventos')
ax2.set_ylabel('Probabilidade')
ax2.grid(True)

# Gráfico de Barras para Distribuição Multinomial
ax3.bar(x_multinomial, multinomial_probs, color='green', alpha=0.7)
ax3.set_title('Distribuição Multinomial')
ax3.set_xlabel('Categoria')
ax3.set_ylabel('Probabilidade')
ax3.grid(True)

# Ajustando a disposição dos subplots
plt.tight_layout()

# Salvando a figura
plt.savefig('distribuicoes_discretas.png', dpi=300)

# Exibindo a figura
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, expon, uniform

# Configuração da Distribuição Normal
mu, sigma = 0, 1
x_norm = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
norm_pdf = norm.pdf(x_norm, mu, sigma)

# Configuração da Distribuição Exponencial
lambda_exp = 1
x_exp = np.linspace(0, 5, 100)
exp_pdf = expon.pdf(x_exp, scale=1/lambda_exp)

# Configuração da Distribuição Uniforme
a, b = 0, 1
x_unif = np.linspace(a, b, 100)
unif_pdf = uniform.pdf(x_unif, a, b-a)

# Criando a figura e os eixos
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))

# Gráfico de Linha para Distribuição Normal
ax1.plot(x_norm, norm_pdf, color='red')
ax1.fill_between(x_norm, norm_pdf, alpha=0.3, color='red')
ax1.set_title('Distribuição Normal')
ax1.set_xlabel('Valor')
ax1.set_ylabel('Densidade de Probabilidade')
ax1.grid(True)

# Gráfico de Linha para Distribuição Exponencial
ax2.plot(x_exp, exp_pdf, color='purple')
ax2.fill_between(x_exp, exp_pdf, alpha=0.3, color='purple')
ax2.set_title('Distribuição Exponencial')
ax2.set_xlabel('Valor')
ax2.set_ylabel('Densidade de Probabilidade')
ax2.grid(True)

# Gráfico de Linha para Distribuição Uniforme
ax3.plot(x_unif, unif_pdf, color='green')
ax3.fill_between(x_unif, unif_pdf, alpha=0.3, color='green')
ax3.set_title('Distribuição Uniforme')
ax3.set_xlabel('Valor')
ax3.set_ylabel('Densidade de Probabilidade')
ax3.grid(True)

# Ajustando a disposição dos subplots
plt.tight_layout()

# Salvando a figura
plt.savefig('distribuicoes_continuas.png', dpi=300)

# Exibindo a figura
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Configuração para a Distribuição Normal
mu, sigma = 0, 1
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
pdf = norm.pdf(x, mu, sigma)
cdf = norm.cdf(x, mu, sigma)

# Criando a figura e os eixos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Gráfico da PDF
ax1.plot(x, pdf, color='blue')
ax1.fill_between(x, pdf, alpha=0.3, color='blue')
ax1.set_title('Função de Densidade de Probabilidade (PDF)')
ax1.set_xlabel('Valor')
ax1.set_ylabel('Densidade de Probabilidade')
ax1.grid(True)

# Gráfico da CDF
ax2.plot(x, cdf, color='red')
ax2.set_title('Função de Distribuição Acumulada (CDF)')
ax2.set_xlabel('Valor')
ax2.set_ylabel('Probabilidade Acumulada')
ax2.grid(True)

# Ajustando a disposição dos subplots
plt.tight_layout()

# Salvando a figura
plt.savefig('pdf_cdf.png', dpi=300)

# Exibindo a figura
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, expon, uniform

def plot_distribution(x_values, y_values, title, color, x_label='Valor', y_label='Densidade de Probabilidade'):
    plt.plot(x_values, y_values, color=color)
    plt.fill_between(x_values, y_values, alpha=0.3, color=color)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True)
    plt.savefig(title + '.png', dpi=300)
    plt.show()

# Configuração da Distribuição Normal
mu, sigma = 0, 1
x_norm = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
norm_pdf = norm.pdf(x_norm, mu, sigma)

# Configuração da Distribuição Exponencial
lambda_exp = 1
x_exp = np.linspace(0, 5, 100)
exp_pdf = expon.pdf(x_exp, scale=1/lambda_exp)

# Configuração da Distribuição Uniforme
a, b = 0, 1
x_unif = np.linspace(a, b, 100)
unif_pdf = uniform.pdf(x_unif, a, b-a)

# Configuração da Distribuição Binomial
n, p = 10, 0.5
x_binom = np.arange(0, n+1)
binom_probs = binom.pmf(x_binom, n, p)

# Configuração da Distribuição de Poisson
mu_poisson = 3
x_poisson = np.arange(0, 10)
poisson_probs = poisson.pmf(x_poisson, mu_poisson)

# Plotando cada distribuição separadamente
plot_distribution(x_norm, norm_pdf, 'Distribuição Normal', 'red')
plot_distribution(x_exp, exp_pdf, 'Distribuição Exponencial', 'purple')
plot_distribution(np.linspace(-10, 10, 100), unif_pdf, 'Distribuição Uniforme', 'green')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom, poisson, multinomial

def plot_distribution(x_values, y_values, title, color, kind='bar', x_label='Valor', y_label='Probabilidade'):
    if kind == 'bar':
        plt.bar(x_values, y_values, color=color, alpha=0.7)
    elif kind == 'line':
        plt.plot(x_values, y_values, 'o-', color=color, markersize=8)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True)
    plt.savefig(title + '.png', dpi=300)
    plt.show()

# Configuração da Distribuição Binomial
n, p = 10, 0.5
x_binom = np.arange(0, n+1)
binom_probs = binom.pmf(x_binom, n, p)

# Configuração da Distribuição de Poisson
mu_poisson = 3
x_poisson = np.arange(0, 10)
poisson_probs = poisson.pmf(x_poisson, mu_poisson)

# Configuração da Distribuição Multinomial
n_multinomial = 10
p_multinomial = [0.2, 0.3, 0.5]
x_multinomial = np.arange(0, n_multinomial+1)
multinomial_probs = multinomial.pmf([3, 3, 4], n_multinomial, p_multinomial)

# Plotando cada distribuição separadamente
plot_distribution(x_binom, binom_probs, 'Distribuição Binomial', 'blue', kind='bar', x_label='Número de Sucessos')
plot_distribution(x_poisson, poisson_probs, 'Distribuição de Poisson', 'blue', kind='line', x_label='Número de Eventos')
plot_distribution(x_multinomial, multinomial_probs, 'Distribuição Multinomial', 'green', kind='bar', x_label='Categoria')
```
